#### 1.데이터 살펴보기


```python
bs_obj = BeautifulSoup(result.content, "html.parser")
# bs_obj
```

```python
#데이터 길이 추출
items= bs_obj.findAll('item')
len(items)
```

```python
#필요 컬럼만 추출하여 df 생성
import pandas as pd
sigungu = [item.signgunm.text for item in items]
signgu_code = [item.signgucode.text for item in items]
day = [item.daywkdivnm.text for item in items]
type_v = [item.toudivnm.text for item in items]
num_v = [item.tounum.text for item in items]
ymd=[item.baseymd.text for item in items]
```

```python
df=pd.DataFrame({'시군구_코드':signgu_code,'시군구':sigungu,'요일':day,'관광객_구분명':type_v,'방문객_수':num_v,'년월일':ymd})
df.head()
```

-추후 합칠 서비스코드 파이레서 시군구 코드가 int 타입
